```python
import pandas as pd
import matplotlib.pyplot as plt

# loading files
roberta_df = pd.read_csv("roberta_predictions.csv")
vader_df = pd.read_csv("vader_predictions.csv")

# merge them into one DataFrame
roberta_df['vader_pred'] = vader_df['vader_pred']

# Count predictions
labels = ['negative', 'neutral', 'positive']
roberta_counts = roberta_df['roberta_pred'].value_counts().reindex(labels, fill_value=0)
vader_counts = roberta_df['vader_pred'].value_counts().reindex(labels, fill_value=0)

# creating the chart
x = range(len(labels))
plt.figure(figsize=(8, 5))
plt.bar([i - 0.15 for i in x], roberta_counts, width=0.3, label='RoBERTa')
plt.bar([i + 0.15 for i in x], vader_counts, width=0.3, label='Custom VADER')
plt.xticks(ticks=x, labels=labels)
plt.ylabel("Number of Predictions")
plt.title("Sentiment Predictions: RoBERTa vs. Custom VADER")
plt.legend()
plt.tight_layout()
plt.show()
```
